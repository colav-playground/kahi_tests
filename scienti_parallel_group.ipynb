```python
from pymongo import MongoClient
import re
from joblib import Parallel, delayed
```

```python
doi="https://10.1172/jci66407"
```

```python
re.sub(r'https*\:\/\/[\w\.]','',doi)
```

```python
re.sub(r'https*\:\/\/[\w\.]+\/','',doi.lower())
```

```python
pipeline = [
     {
       "$match": { "TXT_DOI": { "$ne": None } }
     },
     { "$project": { "doi": { "$trim": { "input": "$TXT_DOI" ,} } } }
     ,
     { "$project": { "doi": { "$toLower": "$doi" } } }
      ,
     { "$group" : { "_id" : "$doi", "ids": { "$push": "$_id" } } }
]
```

```python
db=MongoClient()["scienti_udea_2023"]
```

```python
data = list(db["product"].aggregate(pipeline))
```

```python
data[0]
```

```python
def process_one(reg):
    pass
    
def process_group(group):
    col = MongoClient()["scienti_udea_2023"]["product"]
    for i in group["ids"]:
        reg = col.find_one({"_id":i})
        process_one(reg)            
```

```python
Parallel(
    n_jobs=10,
    verbose=10,
    backend="threading")(
    delayed(process_group)(group) for group in data
)
```
